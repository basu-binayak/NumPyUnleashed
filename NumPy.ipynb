## A. Creating NumPy Arrays


### A.1 Creating NumPy Arrays from List 

```python
# Integer array 
int_array = np.array([1,5,6,8,9,3])
print(int_array)

# float array 
# note that numpy arrays are homogenous - hence if some types don't match, numpy will upcast them according to the promotion rules. In the example below, an int in upcast to float
float_array = np.array([2.3,4.5,6.7,5,8.9])
print(float_array)

# You can also mention the type while crearing the array 
float_array = np.array([1,2,3,4,5], dtype=np.float32)
print(float_array)

# Python lists are always one dimensional - however NumPy arrays can be multidimensional 
md_array = np.array([range(j,j+5) for j in range(1,4)]) # this is a 3x5 array
print(md_array)
```

### A.2 Creating NumPy Arrays from Scratch

---

#### üîç Difference Between 1D and 2D Arrays in NumPy

---

#### üü¶ **1D Array (One-Dimensional)**

A 1D array is like a **simple list of values** ‚Äî all arranged in a single row (or line).


#### üü® **2D Array (Two-Dimensional)**

A 2D array is like a **table or matrix** ‚Äî it has **rows and columns**.

#### üß† Key Differences Summary

| Feature        | 1D Array                        | 2D Array                              |
|----------------|----------------------------------|----------------------------------------|
| Structure      | Single list                     | List of lists (rows & columns)        |
| Shape          | `(n,)`                          | `(rows, cols)`                        |
| Visualization  | Horizontal line of values       | Table/grid                            |
| Real-world Use | Vectors, signals                | Matrices, images, spreadsheets        |

---

#### üìä Table: Methods to Create NumPy Arrays from Scratch

| Method         | Description                            | 1D Array Syntax                            | 2D Array Syntax                                  |
|----------------|----------------------------------------|--------------------------------------------|--------------------------------------------------|
| `np.zeros()`   | Creates an array filled with `0`s      | `np.zeros(5)`                              | `np.zeros((2, 3))`                               |
| `np.ones()`    | Creates an array filled with `1`s      | `np.ones(4)`                               | `np.ones((3, 2))`                                |
| `np.full()`    | Creates an array filled with a value   | `np.full(4, 7)`                            | `np.full((2, 3), 9)`                             |
| `np.arange()`  | Creates an array with a range of values| `np.arange(0, 10, 2)`                      | `np.arange(1, 7).reshape((2, 3))`                |
| `np.linspace()`| Creates evenly spaced values           | `np.linspace(0, 1, 5)`                     | `np.linspace(0, 10, 6).reshape((2, 3))`          |

---

#### ‚úÖ Notes

- `np.full(shape, fill_value)` lets you fill arrays with any number.
- `np.arange(start, stop, step)` generates values like `range()`.
- `np.linspace(start, stop, num)` returns `num` evenly spaced points.
- For 2D arrays using `arange` or `linspace`, use `.reshape(rows, cols)`.

---

#### `np.zeros()`

```python
# np.zeros()

# 1D array
print(np.zeros(8)) # by default, creates a floating array of 0s of length 8
print(np.zeros(8, dtype=np.int32)) # create a integer array now 

# multidimensional
print(np.zeros((3,5))) # provide dtype if you want specifically a data type 
```

#### `np.ones()`

```python
# np.ones()

# 1D array
print(np.ones(8))

# multidimensional array
print(np.ones((4,5)))
```

#### `np.full()`

```python
# np.full()

# 1D array
print(np.full(5, fill_value=2.3)) # creates a array of length 5 filled with 2.3

# multidimensional array
print(np.full((3,5), fill_value=3.4))
```

#### `np.arange()`

```python
# np.arange 

print(np.arange(start=0, stop=20, step=2.5)) # this creates a array that starts at 0, goes till 20 with a step size of 2.5

print(np.arange(start=0, stop=20).reshape(5,4)) # use reshape() method to reshape the 1D array to a 2D array 
```

#### `np.linspace()`

```python
# np.linspace 

print(np.linspace(start=0, stop=20, num=30)) # gives a array of length 30 starting from 0 and ending at 20 equally spaced

print(np.linspace(start=0, stop=20, num=30).reshape(6,5)) # convert the 1D to a 2D array
```

#### `np.eye()`

```python
np.eye(3) # creates a identity matrix of 3x3
```

### A.3 Creating Pseudo-random NumPy Arrays
---

#### üé≤ Table: Methods to Create Pseudo-Random Arrays in NumPy

| Method                          | Description                                    | 1D Array Syntax                              | 2D Array Syntax                                  |
|----------------------------------|------------------------------------------------|----------------------------------------------|--------------------------------------------------|
| `np.random.rand()`              | Random floats in [0.0, 1.0)                    | `np.random.rand(5)`                          | `np.random.rand(2, 3)`                           |
| `np.random.randn()`             | Random samples from standard normal dist.      | `np.random.randn(4)`                         | `np.random.randn(3, 2)`                          |
| `np.random.randint()`           | Random integers in a range                     | `np.random.randint(0, 10, size=5)`           | `np.random.randint(0, 100, size=(2, 3))`         |
| `np.random.uniform()`           | Random floats from a uniform distribution      | `np.random.uniform(1, 5, size=4)`            | `np.random.uniform(10, 20, size=(3, 2))`         |
| `np.random.normal()`            | Random samples from a normal distribution      | `np.random.normal(0, 1, size=5)`             | `np.random.normal(10, 2, size=(2, 3))`           |
| `np.random.choice()`            | Randomly pick elements from a given list       | `np.random.choice([1, 2, 3, 4], size=5)`     | `np.random.choice([10, 20], size=(2, 3))`        |


#### ‚úÖ Quick Notes:

- `rand` and `randn` are **quick shorthands** (no keyword `size=`, just shape directly).
- `randint(low, high, size)` gives **integers**, unlike `uniform` and `normal` which give **floats**.
- `choice` is great for **simulating dice rolls**, **lotteries**, or **categorical picks**.
- All methods are **pseudo-random** (deterministic if you set a seed with `np.random.seed()`).

---

#### `np.random.rand()`

```python
# np.random.rand()

print(np.random.rand(3,3))
print(np.random.rand(3))
```

```python
# check the distribution
import matplotlib.pyplot as plt 
import seaborn as sns
sns.set_style('darkgrid')

plt.figure(figsize=(12,6))
sns.histplot(np.random.rand(10000), kde=True, color='green')

plt.show()
```

#### `np.random.randn()`

```python
# np.random.randn
print(np.random.randn(3,3))
print(np.random.randn(3))
```

```python
# check the standard normal distribution
plt.figure(figsize=(12,6))
sns.histplot(np.random.randn(10000), kde=True, color='green')

plt.show()
```

#### `np.random.randint()`

```python
# Create a 3x3 array of  pseudorandom integer values between 2 and 20
print(np.random.randint(low=2, high=20, size=(3,3)))

# 1D array
print(np.random.randint(low=2, high=20, size=9))
```

```python
# check the distribution

plt.figure(figsize=(12,6))
sns.histplot(np.random.randint(0,100,10000), kde=True, color='green')

plt.show()
```

#### `np.random.uniform()`

```python
np.random.uniform(low=1, high=50, size=10) # 1D array
```

```python
# check the distribution

plt.figure(figsize=(12,6))
sns.histplot(np.random.uniform(0,100,10000), kde=True, color='green')

plt.show()
```

#### `np.random.normal()`

```python
# 1D array
# loc - mean, scale - std
np.random.normal(loc=1, scale=2, size=10)
```

```python
# check the distribution

plt.figure(figsize=(12,6))
sns.histplot(np.random.normal(0,1,10000), kde=True, color='green')

plt.show()
```

#### `np.random.choice()`

```python
random_arr = np.random.normal(0,1,100000)

# select 10 random values from the random_arr
np.random.choice(random_arr, size=10)
```

```python
# check the distribution

plt.figure(figsize=(12,6))
sns.histplot(np.random.choice(random_arr, size=10000), kde=True, color='green')

plt.show()
```
